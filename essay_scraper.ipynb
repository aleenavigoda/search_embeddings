```python
def scrape_essay_content(target_url):
    try:
        # Fetch the webpage content
        response = requests.get(target_url)
        if response.status_code != 200:
            print(f"Failed to retrieve {target_url}")
            return None
        
        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Try to find the main content of the essay
        # Adjust this based on the actual structure of the essays you are scraping
        main_content = soup.find('article') or soup.find('main') or soup.find('div', class_="essay-content")
        
        if main_content:
            # Get the text from the found content
            return main_content.get_text(separator=" ", strip=True)
        else:
            # If no specific tag is found, fallback to extracting all paragraphs
            paragraphs = soup.find_all('p')
            essay_text = ' '.join(p.get_text(separator=" ", strip=True) for p in paragraphs)
            return essay_text if essay_text else None
    except Exception as e:
        print(f"Error scraping {target_url}: {e}")
        return None
```

```python
import requests
from bs4 import BeautifulSoup
import time
from urllib.parse import urlparse
from collections import defaultdict

def scrape_essay_content(target_url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        
        response = requests.get(target_url, headers=headers, timeout=10)
        response.raise_for_status()
        
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Try multiple methods to find the main content
        content = None
        
        # Method 1: Look for specific article or content divs
        content = soup.find('article') or soup.find('div', class_="content") or soup.find('div', class_="post-content")
        
        # Method 2: If not found, try to get all paragraph tags
        if not content:
            content = soup.find_all('p')
        
        # Method 3: If still not found, get the body content
        if not content:
            content = soup.find('body')
        
        if content:
            if isinstance(content, list):
                # If it's a list of elements, join their text
                text = ' '.join([elem.get_text(strip=True) for elem in content])
            else:
                # If it's a single element, get its text
                text = content.get_text(separator=" ", strip=True)
            
            # Remove extra whitespace
            text = ' '.join(text.split())
            
            print(f"Scraped {len(text)} characters from {target_url}")
            return text
        else:
            print(f"No content found for {target_url}")
            return None
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {target_url}: {e}")
    except Exception as e:
        print(f"Error processing {target_url}: {e}")
    return None

def get_domain(url):
    return urlparse(url).netloc

def scrape_and_update_by_domain():
    try:
        # Fetch all rows from the urls_table
        response = supabase.table('urls_table').select('id', 'url').execute()
        essays = response.data

        # Group essays by domain
        domain_groups = defaultdict(list)
        for essay in essays:
            domain = get_domain(essay['url'])
            domain_groups[domain].append(essay)

        # Process each domain group
        for domain, domain_essays in domain_groups.items():
            print(f"Processing domain: {domain}")
            for essay in domain_essays:
                essay_id = essay['id']
                url = essay['url']
                
                scraped_text = scrape_essay_content(url)
                
                if scraped_text:
                    update_response = supabase.table('urls_table').update({"full_text": scraped_text}).eq('id', essay_id).execute()
                    print(f"Updated full_text for essay_id {essay_id}. URL: {url}")
                else:
                    print(f"No text scraped for essay_id {essay_id}. URL: {url}")
                
                time.sleep(2)  # Delay between requests to the same domain
            
            print(f"Finished processing domain: {domain}")
            time.sleep(5)  # Longer delay between different domains

    except Exception as e:
        print(f"Error in scrape_and_update_by_domain: {e}")

# Run the function to scrape and update essays by domain
scrape_and_update_by_domain()
```
